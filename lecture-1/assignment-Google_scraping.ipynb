```python
import sqlite3
```

# データベースの作成

```python
path = ''
db_name = 'google_repository.db'

try:
    conn = sqlite3.connect(path + db_name)
    cur = conn.cursor()

    #SQL文の作成 (CREATE TABLE)
    sql = 'CREATE TABLE repository (name TEXT, main_language TEXT ,stars INTEGER);'

    cur.execute(sql)

except sqlite3.Error as e:
    print('エラーが発生しました',e)
    
finally:
    conn.close()
```

# 　スクレイピングでリストに保存

```python
#必要なライブラリのインポート
import requests
from bs4 import BeautifulSoup
import time
```

```python
repos_info = []
page = 1
while page < 95:
    url = f"https://github.com/orgs/google/repositories?page={page}"
    print(f'{page}ページ目')
    
    res = requests.get(url)
    res.encoding = res.apparent_encoding
    soup = BeautifulSoup(res.text,"html.parser")
    repos = soup.find_all(class_='ListItem-module__listItem--k4eMk')
    try:
         for repo in repos:
                try:
                    name = repo.find(class_='Title-module__anchor--GmXUE Title-module__inline--oM0P7')
                    language = repo.find(class_='ReposListItem-module__Text_4--mkG7R')
                    stars = repo.select_one('a[href$="/stargazers"]') #forks数と混同してしまうためCSSセレクタで検索

                    if name:
                        name = name.text.strip()
                        try:
                            language = language.text.strip()
                        except AttributeError:
                            language = None
                              
                        stars_text = None
                        stars_count = 0
                        if stars:
                            # テキスト→数字のみ（kあり）
                            stars_text = stars.get_text().strip()
                            
                            try:
                                if 'k' in stars_text:
                                    stars_count = int(float(stars_text.replace('k', '')) * 1000)
                                else:
                                    # kなし
                                    stars_count = int(float(stars_text))
                            except ValueError:
                                stars_count = 0
                        
                        repos_info.append((name,language,stars_count))
                        print(name,language,stars_count)
                          
                except Exception as e:
                    print("エラー:",e)       
                    
                time.sleep(1)

    except Exception as e:
            print('エラーが発生しました',e)
            
    finally:
            print(f'{page}ページ目終了')
    page += 1
            
print('ウェブスクレイピング処理が完了しました。')
```

# データベースにinsertする

```python
path = ''
db_name = 'google_repository.db'

try:
    conn = sqlite3.connect(path + db_name)
    cur = conn.cursor()
    
    #SQL文(INSERT)
    sql = "INSERT INTO repository (name , main_language , stars) VALUES(?, ?, ?);"
    #実行
    cur.executemany(sql,repos_info)
    #変更を反映
    conn.commit()

except sqlite3.Error as e:
    print('エラーが発生しました',e)
    
finally:
    conn.close()
```

# データをSELECT文で表示する

```python
path = ''
db_name = 'google_repository.db'

try:
    conn = sqlite3.connect(path + db_name)
    cur = conn.cursor()

    #SQL文の作成 
    sql = "SELECT * FROM  repository ;"

    #SQl文の実行
    cur.execute(sql)

except sqlite3.Error as e:
    print('エラーが発生しました',e)
    
else:
    for row in cur:
        name, main_language,stars = row
        print(name, main_language,stars)    
    
finally:
    conn.close()
```
